```python
import torch
import matplotlib.pyplot as plt

metrics = torch.load("../task_3/lrs.pth")
lrs = metrics["lrs"]
losses = metrics["losses"]

plt.plot(lrs, losses)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Learning Rate")
plt.ylabel("Loss")
plt.title("Learning Rate vs Loss")
plt.show()
```
